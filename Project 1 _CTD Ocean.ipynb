# Axial Base Deep Profiler

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("Winterdata.csv")
data.shape
T=data.temp                     #Temperature
S=data.practical_salinity       #Salinity
z=data.pressure                 #Height
t=data.time
y=c=1449.2+4.6*(T)-0.055*(T)**2+(1.34-.01*T)*(S-35)+0.016*z  #Speed of Sound
x=z=data.pressure

plt.plot(x,y)
plt.ylabel('Speed of Sound (ssp)')
plt.xlabel('Depth [dbar]')
plt.title('Depth(m) vs ssp(m/s) (Axial Deep Jan1-2)')

#plot the data
plt.plot(x,y,'b',alpha=0.3,ms=4,label='data')
    #plt.xlabel('Time')
    #plt.ylabel('Intensity')
#define a moving average function
def moving_average(x,y,step_size=.1,bin_size=1):
    bin_centers  = np.arange(np.min(x),np.max(x)-0.5*step_size,step_size)+0.5*step_size
    bin_avg = np.zeros(len(bin_centers))
    for index in range(0,len(bin_centers)):
        bin_center = bin_centers[index]
        items_in_bin = y[(x>(bin_center-bin_size*0.5) ) & (x<(bin_center+bin_size*0.5))]
        bin_avg[index] = np.mean(items_in_bin)
    return bin_centers,bin_avg
#plot the moving average
bins, average = moving_average(x,y)
plt.plot(bins, average,label='moving average')
plt.legend()
plt.show()

def getnumberofDives(z): 
    bottom = max(z)
    atbottom = False
    divecount=0
    for i in range(0,len(z)):
        if(z[i]>0.95*bottom and not atbottom):
            atbottom=True
        if (z[i]<0.95*bottom and atbottom):
            atbottom=False
            divecount+=1
    return divecount
N= getnumberofDives(z)
print(N, "dives")
```
